```
import pandas as pd
import numpy as np
```

```
ds=pd.read_csv('breast_cancer.csv')
x=ds.iloc[:,1:-1].values
y=ds.iloc[:,-1].values
ds.head()
```

```
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=0)
```

```
from sklearn.linear_model import LogisticRegression
logc=LogisticRegression(random_state=0)
logc.fit(xtrain,ytrain)
```

```
pred=logc.predict(xtest)
print('beningn=2;malignant=4')
print(np.concatenate((pred.reshape(len(pred),1),ytest.reshape(len(ytest),1)),1))
```

```
from sklearn.metrics import confusion_matrix
print(confusion_matrix(ytest,pred))
(103+59)/(103+59+4+5)
```
